```python
import pandas as pd
from tqdm import tqdm
from transformers import pipeline

def analyze_reviews(reviews):
    # ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
    model_name = "MLOpsEngineer/review-rate-prediction"
    classifier = pipeline("sentiment-analysis", model=model_name)

    results = []
    for review_text in tqdm(reviews):
        # ëª¨ë¸ ì¶”ë¡ 
        result = classifier(review_text)[0]
        # ê²°ê³¼ í•´ì„
        label = result['label']
        score = result['score']

        if label in ['LABEL_0', 'LABEL_1']:
            sentiment = 0 # 'ë¶€ì •'
        else:
            sentiment = 1 # 'ê¸ì •'

        results.append((review_text, sentiment, score))

    # ë°ì´í„°í”„ë ˆì„ ìƒì„±
    tmp = pd.DataFrame(results, columns=['review', 'sentiment', 'score'])
    return tmp
```

```python
from sklearn.cluster import KMeans

# K-means í´ëŸ¬ìŠ¤í„°ë§ ëª¨ë¸ ìƒì„±
kmeans = KMeans(n_clusters=5)  # êµ°ì§‘ì˜ ê°œìˆ˜ë¥¼ ì§€ì • (ì˜ˆ: 3ê°œì˜ êµ°ì§‘)

# êµ°ì§‘í™”ì— ì‚¬ìš©í•  ë°ì´í„° ì„ íƒ
X = final_result_id_df2[['MEAN_NEW_SCORE', 'id_count']]

# K-means í´ëŸ¬ìŠ¤í„°ë§ ëª¨ë¸ í›ˆë ¨
kmeans.fit(X)

# ê° ë°ì´í„° í¬ì¸íŠ¸ì— ëŒ€í•œ êµ°ì§‘ ë ˆì´ë¸” ì˜ˆì¸¡
labels = kmeans.predict(X)

# êµ°ì§‘í™” ê²°ê³¼ë¥¼ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
final_result_id_df2['Cluster'] = labels

# êµ°ì§‘í™” ê²°ê³¼ë¥¼ ì‹œê°í™”í•œ scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MEAN_NEW_SCORE', y='id_count', hue='Cluster', data=final_result_id_df2, palette='Set1')

# ê·¸ë˜í”„ ì œëª©ê³¼ ë ˆì´ë¸” ì„¤ì •
plt.title('Clustered Relation between Review Temperature & Review Frequency')
plt.xlabel('Review Temperature')
plt.ylabel('Review Frequency')

# ê·¸ë˜í”„ í‘œì‹œ
plt.tight_layout()
plt.show()
```
