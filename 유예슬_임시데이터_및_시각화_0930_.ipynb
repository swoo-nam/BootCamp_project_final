```python
import pandas as pd
from tqdm import tqdm
from transformers import pipeline

def analyze_reviews(reviews):
    # 모델 불러오기
    model_name = "MLOpsEngineer/review-rate-prediction"
    classifier = pipeline("sentiment-analysis", model=model_name)

    results = []
    for review_text in tqdm(reviews):
        # 모델 추론
        result = classifier(review_text)[0]
        # 결과 해석
        label = result['label']
        score = result['score']

        if label in ['LABEL_0', 'LABEL_1']:
            sentiment = 0 # '부정'
        else:
            sentiment = 1 # '긍정'

        results.append((review_text, sentiment, score))

    # 데이터프레임 생성
    tmp = pd.DataFrame(results, columns=['review', 'sentiment', 'score'])
    return tmp
```

```python
from scipy.stats import ttest_ind

# 음수 평균 점수와 양수 평균 점수를 각각의 그룹으로 나누기
negative_scores = final_result_id_df2[final_result_id_df2['MEAN_NEW_SCORE'] < 0]['id_count']
positive_scores = final_result_id_df2[final_result_id_df2['MEAN_NEW_SCORE'] >= 0]['id_count']

# t-검정 수행
t_stat, p_value = ttest_ind(negative_scores, positive_scores)

# 결과 출력
print(f't-검정 통계량: {t_stat}')
print(f'p-value: {p_value}')

# p-value를 기준으로 유의미성 판단
alpha = 0.05
if p_value < alpha:
    print('음수 평균 점수와 양수 평균 점수 간의 차이는 유의미합니다.')
else:
    print('음수 평균 점수와 양수 평균 점수 간의 차이는 유의미하지 않습니다.')
```
