```python
def review_topic_analysis(reviews, threshold=0.3):
    # 주제별 쿼리 설정
    queries = ["Taste", "Taste","Quantity", "Price", "Service", "Delivery"]

    # 주제별 관련성 점수를 저장할 리스트
    scores_by_topic = []

    # 모델과 토크나이저 로드
    model_name = "nboost/pt-bert-large-msmarco"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)

    # 각 주제별로 모델 예측 수행
    for query in queries:
        # 리뷰와 주제별 쿼리를 토큰화
        inputs = tokenizer(reviews, [query] * len(reviews), return_tensors="pt", padding=True, truncation=True, max_length=512)
        # 모델 예측
        with torch.no_grad():
            logits = model(**inputs).logits
        # 관련성 점수 추출
        scores = logits[:, 1].tolist()
        scores_by_topic.append(scores)

    # 각 리뷰에 대해 관련된 주제들을 선택
    related_topics = []
    for i in range(len(reviews)):
        topic_scores = [(topic[i], idx) for idx, topic in enumerate(scores_by_topic)]
        topic_scores.sort(reverse=True, key=lambda x: x[0])  # 점수를 기준으로 내림차순 정렬
        if topic_scores[0][0] - topic_scores[1][0] < threshold:
            related_topics.append(f"{queries[topic_scores[0][1]]}, {queries[topic_scores[1][1]]}")
        else:
            related_topics.append(queries[topic_scores[0][1]])

    # 결과를 데이터프레임으로 저장
    data = {'리뷰': reviews}
    for i, query in enumerate(queries):
        data[f"{query} 관련도"] = scores_by_topic[i]
    data['관련 주제들'] = related_topics

    df = pd.DataFrame(data)

    return df


df = review_topic_analysis(translated_reviews)
df
```

```python
# z코드 개선

def review_topic_analysis(reviews, threshold=0.5):
    # 주제별 쿼리 설정
    queries = ["Food", "Quantity", "Price", "Service", "Delivery"]

    # 모든 쿼리에 대한 입력을 한 번에 처리하기 위해 토큰화
    all_queries = [q for q in queries for _ in reviews]
    repeated_reviews = reviews * len(queries)

    inputs = tokenizer(repeated_reviews, all_queries, return_tensors="pt", padding=True, truncation=True, max_length=512)

    # 모델 예측
    with torch.no_grad():
        logits = model(**inputs).logits

    # 각 리뷰에 대한 점수 분할 및 재구성
    scores_by_topic = [logits[i::len(reviews), 1].tolist() for i in range(len(queries))]

    # 각 리뷰에 대해 관련된 주제들 선택
    related_topics = []
    for i in range(len(reviews)):
        scores_for_review = [topic[i] for topic in scores_by_topic]
        topic_scores = sorted([(score, idx) for idx, score in enumerate(scores_for_review)], reverse=True)

        # 첫 번째 주제와 두 번째 주제의 점수 차이 확인
        if topic_scores[0][0] - topic_scores[1][0] < threshold:
            # 첫 번째와 두 번째 주제의 점수 차이가 임계값보다 작은 경우 세 번째 주제의 점수도 확인
            if topic_scores[1][0] - topic_scores[2][0] < threshold:
                sum_of_top_three = topic_scores[0][0] + topic_scores[1][0] + topic_scores[2][0]
                first_percentage = (topic_scores[0][0] / sum_of_top_three) * 100
                second_percentage = (topic_scores[1][0] / sum_of_top_three) * 100
                third_percentage = (topic_scores[2][0] / sum_of_top_three) * 100
                related_topics.append(f"{queries[topic_scores[0][1]]} ({first_percentage:.2f}%), {queries[topic_scores[1][1]]} ({second_percentage:.2f}%), {queries[topic_scores[2][1]]} ({third_percentage:.2f}%)")
            else:
                sum_of_top_two = topic_scores[0][0] + topic_scores[1][0]
                first_percentage = (topic_scores[0][0] / sum_of_top_two) * 100
                second_percentage = (topic_scores[1][0] / sum_of_top_two) * 100
                related_topics.append(f"{queries[topic_scores[0][1]]} ({first_percentage:.2f}%), {queries[topic_scores[1][1]]} ({second_percentage:.2f}%)")
        else:
            related_topics.append(f"{queries[topic_scores[0][1]]} (100%)")

    # 결과를 데이터프레임으로 저장
    data = {'리뷰': reviews}
    for i, query in enumerate(queries):
        data[f"{query} 관련도"] = scores_by_topic[i]
    data['관련 주제들'] = related_topics

    df = pd.DataFrame(data)

    return df

# 최적화된 함수 코드를 제공합니다.

df = review_topic_analysis(translated_reviews)
df
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch
from torch.utils.data import Dataset, DataLoader

class ReviewDataset(Dataset):
    def __init__(self, reviews, queries):
        self.reviews = reviews
        self.queries = queries

    def __len__(self):
        return len(self.reviews) * len(self.queries)

    def __getitem__(self, idx):
        review_idx = idx // len(self.queries)
        query_idx = idx % len(self.queries)
        return self.reviews[review_idx], self.queries[query_idx]

def collate_fn(batch):
    reviews, queries = zip(*batch)
    return tokenizer(list(reviews), list(queries), return_tensors="pt", padding=True, truncation=True, max_length=512)


def review_topic_analysis(reviews, threshold=0.3):

    model_name = "nboost/pt-bert-large-msmarco"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)


    # 주제별 세부 쿼리 설정
    topic_queries = {
        "Taste": [
            "delicious", "tasteless", "salty", "sweet", "sour",
            "bitter", "spicy", "flavorful", "rich", "mild",
            "tangy", "savory", "creamy", "crispy", "tender",
            "juicy", "overcooked", "undercooked", "fresh", "stale"
        ],
        "Quantity": [
            "generous", "small portion", "sufficient", "abundant", "scarce",
            "filling", "satisfying", "plenty", "minimal", "hearty",
            "oversized", "undersized", "bountiful", "limited", "overflowing",
            "sparse", "enough", "more than enough", "just right", "not enough"
        ],
        "Price": [
            "good value", "expensive", "cheap", "worth it", "overpriced",
            "budget-friendly", "costly", "fair price", "high-end", "low-end",
            "reasonable", "unreasonable", "affordable", "steep", "moderate",
            "inexpensive", "pricey", "economical", "value for money", "rip-off"
        ],
        "Service": [
            "good response", "rude service", "kind", "friendly", "unfriendly",
            "welcoming", "cold", "professional", "slow",
            "prompt", "dismissive", "respectful", "inattentive", "helpful",
            "unhelpful", "courteous", "warm", "chatty"
        ],
        "Delivery": [
            "fast delivery", "late arrival", "slow delivery", "on-time", "delayed",
            "early", "timely", "postponed", "speedy", "held up",
            "punctual", "dragged", "swift", "prolonged", "ahead of schedule",
            "lagged", "quick", "expedited"
        ]
    }
    all_queries = [sub_query for queries in topic_queries.values() for sub_query in queries]

    dataset = ReviewDataset(reviews, all_queries)

    dataloader = DataLoader(dataset, batch_size=16, shuffle=False, collate_fn=collate_fn)

    all_scores = []
    with torch.no_grad():
        for inputs in dataloader:
            inputs = {key: val.to(device) for key, val in inputs.items()}
            logits = model(**inputs).logits
            scores = torch.softmax(logits, dim=-1)[:, 1].tolist()
            all_scores.extend(scores)

    # 모든 쿼리에 대한 입력을 한 번에 처리하기 위해 토큰화
    inputs = {key: val.to(device) for key, val in tokenizer(reviews * len(all_queries), all_queries * len(reviews), return_tensors="pt", padding=True, truncation=True, max_length=512).items()}

    # 모델 예측
    with torch.no_grad():
        logits = model(**inputs).logits

    # 각 리뷰에 대한 점수 분할 및 재구성
    scores_by_topic = [logits[i::len(reviews), 1].tolist() for i in range(len(all_queries))]

    # 주제별 점수 합산
    aggregated_scores = {}
    idx = 0
    for topic, sub_queries in topic_queries.items():
        aggregated_scores[topic] = [sum(scores_by_topic[idx + j][k] for j in range(len(sub_queries))) for k in range(len(reviews))]
        idx += len(sub_queries)

    related_topics = []
    for i in range(len(reviews)):
        scores_for_review = [topic[i] for topic in aggregated_scores.values()]
        topic_scores = sorted([(score, idx) for idx, score in enumerate(scores_for_review)], reverse=True)

        # 첫 번째 주제와 두 번째 주제의 점수 차이 확인
        if topic_scores[0][0] - topic_scores[1][0] < threshold:
            # 첫 번째와 두 번째 주제의 점수 차이가 임계값보다 작은 경우 세 번째 주제의 점수도 확인
            if topic_scores[1][0] - topic_scores[2][0] < threshold:
                sum_of_top_three = topic_scores[0][0] + topic_scores[1][0] + topic_scores[2][0]
                first_percentage = (topic_scores[0][0] / sum_of_top_three) * 100
                second_percentage = (topic_scores[1][0] / sum_of_top_three) * 100
                third_percentage = (topic_scores[2][0] / sum_of_top_three) * 100
                related_topics.append(f"{list(topic_queries.keys())[topic_scores[0][1]]} ({first_percentage:.2f}%), {list(topic_queries.keys())[topic_scores[1][1]]} ({second_percentage:.2f}%), {list(topic_queries.keys())[topic_scores[2][1]]} ({third_percentage:.2f}%)")
            else:
                sum_of_top_two = topic_scores[0][0] + topic_scores[1][0]
                first_percentage = (topic_scores[0][0] / sum_of_top_two) * 100
                second_percentage = (topic_scores[1][0] / sum_of_top_two) * 100
                related_topics.append(f"{list(topic_queries.keys())[topic_scores[0][1]]} ({first_percentage:.2f}%), {list(topic_queries.keys())[topic_scores[1][1]]} ({second_percentage:.2f}%)")
        else:
            related_topics.append(f"{list(topic_queries.keys())[topic_scores[0][1]]} (100%)")

    # 결과를 데이터프레임으로 저장
    data = {'Review': reviews, 'Related Topics': related_topics}
    for topic, scores in aggregated_scores.items():
        data[f"{topic} Score"] = scores

    df = pd.DataFrame(data)

    return df

df = review_topic_analysis(translated_reviews)
df
```
