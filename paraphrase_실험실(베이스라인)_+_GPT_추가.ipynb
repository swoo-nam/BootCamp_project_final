````python
# 가게명 패턴 함수들
def text_ver_1():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '저희\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "저희" 없어서 패스')
            pass
        else:
            tmp = re.split('의', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                elif t[0] == '매':
                    # print(f'{i}번 "가게명" 없어서 패스')
                    break
                else:
                    t = t + '의'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_1 실행완료')


def text_ver_2():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '안녕하세요\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "안녕하세요" 없어서 패스')
            pass
        else:
            tmp = re.split('입니다', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                else:
                    t = t + '입니다'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_2 실행완료')


def text_ver_3():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '안녕하세요.\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "안녕하세요." 없어서 패스')
            pass
        else:
            tmp = re.split('입니다.', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                else:
                    t = t + '입니다.'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_3 실행완료')


def text_ver_4():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '저희\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "저희" 없어서 패스')
            pass
        else:
            tmp = re.split('을', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                elif t[0] == '매':
                    # print(f'{i}번 "가게명" 없어서 패스')
                    break
                else:
                    t = ' ' + t + '을'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '를')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_4 실행완료')


def text_ver_5():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '저 희\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "저희" 없어서 패스')
            pass
        else:
            tmp = re.split('을', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                elif t[0] == '매':
                    # print(f'{i}번 "가게명" 없어서 패스')
                    break
                else:
                    t = ' ' + t + '을'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '를')
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace('저 희', '저희')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_5 실행완료')


def text_ver_6():
    for i in tqdm(range(yogi_df['replycontent'].shape[0])):
        pat = '안녕하세요!\s+'
        tmp = re.split(pat, yogi_df['replycontent'][i])
        if len(tmp) == 1:
            # print(f'{i}번 "안녕하세요." 없어서 패스')
            pass
        else:
            tmp = re.split('입니다!', tmp[1])
            cnt = 0
            for t in tmp:
                if len(t) == 0:
                    pass
                elif t[0] == ' ':
                    # print(f'{i}번 수정완료')
                    break
                else:
                    t = t + '입니다!'
                    yogi_df['replycontent'][i] = yogi_df['replycontent'][i].replace(t, '')
                    cnt += 1
                    if cnt == len(tmp):
                        # print(f'{i}번 수정부분 없음')
                        pass

    print('text_ver_6 실행완료')


# 패턴 함수들 실행시킬 함수
def text_remove(*args):
    if len(args) == 0:
        text_ver_1()
        text_ver_2()
        text_ver_3()
        text_ver_4()
        text_ver_5()
        text_ver_6()
    elif len(args) > 1:
        print('숫자를 1개만 입력해주세요.')
    elif args[0] == 1:
        text_ver_1()
    elif args[0] == 2:
        text_ver_2()
    elif args[0] == 3:
        text_ver_3()
    elif args[0] == 4:
        text_ver_4()
    elif args[0] == 5:
        text_ver_5()
    elif args[0] == 6:
        text_ver_6()
    else:
        print('1 ~ 6 사이의 숫자를 입력해주세요')
````

````python
# 불필요한 문자 제거

# 고객리뷰 부분
content_list = []
for content in yogi_df['content']:
    if isinstance(content, str):
        c_result = re.sub(r'\,{2,}', ',', content)
        c_result = re.sub(r'\~{2,}', '~', c_result)
        c_result = re.sub(r'\!{2,}', '!', c_result)
        c_result = re.sub(r'\?{2,}', '?', c_result)
        # c_result = re.sub(r'\.{2,}', '.', c_result)
        c_result = re.sub(' +', ' ', c_result)
        c_result = c_result.lstrip()

        content_list.append(c_result)
    else:
        content_list.append('')

yogi_df['content'] = content_list

# 사장댓글 부분
replyContent_list = []
for replyContent in yogi_df['replycontent']:
    if isinstance(replyContent, str):
        r_result = re.sub(r'\,{2,}', ',', replyContent)
        r_result = re.sub(r'\~{2,}', '~', r_result)
        r_result = re.sub(r'\!{2,}', '!', r_result)
        r_result = re.sub(r'\?{2,}', '?', r_result)
        # r_result = re.sub(r'\.{2,}', '.', r_result)
        r_result = re.sub(r' +', ' ', r_result)
        r_result = r_result.lstrip()

        replyContent_list.append(r_result)
    else:
        replyContent_list.append('')

yogi_df['replycontent'] = replyContent_list

yogi_df.head()
````

````python
model_name = 'skt/kogpt2-base-v2'
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM
````

````python
model = AutoModelForCausalLM.from_pretrained(model_name)
````

````python
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          bos_token='</s>',
                                          eos_token='</s>',
                                          unk_token='<unk>',
                                          pad_token='<pad>',
                                          mask_token='<mask>',
                                          max_len=1024)
````

````python
class ChatDataset(torch.utils.data.Dataset):
    def __init__(self, df):
        self.content = shuffled_df['content'].tolist()
        self.replyContent = shuffled_df['replycontent'].tolist()

    def __len__(self):
        return len(self.content)

    def __getitem__(self, idx):
        return '<q>' + self.content[idx] + '</s><a>' + self.replyContent[idx] + '</s>'
````

````python
def collate_fn(batch):
    x = tokenizer(batch, return_tensors='pt', padding=True)
    return {'x': x}
````

````python
dt = ChatDataset(shuffled_df)
dl = torch.utils.data.DataLoader(dt, batch_size=2, collate_fn=collate_fn)
batch = next(iter(dl))
batch
````

````python
model(**batch['x']).logits.shape
````

````python
tokenizer.pad_token_id
````

````python
def train_loop(dataloader, model, loss_fn, optimizer, device):
    epoch_loss = 0
    model.train()

    for batch in tqdm(dataloader):
        x = batch['x'].to(device)
        pred = model(**x).logits
        n_class = pred.shape[-1]
        pred = pred[:,:-1]
        pred = pred.reshape(-1, n_class)

        tgt = x['input_ids'][:,1:]
        tgt = tgt.flatten()

        mask = tgt != 3
        tgt = tgt[mask]
        pred = pred[mask]
        loss = loss_fn(pred,tgt)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    epoch_loss /= len(dataloader)

    return epoch_loss
````

````python
reset_seeds(SEED)

model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=3e-5)

train_dt = ChatDataset(train_df)
train_dl = torch.utils.data.DataLoader(train_dt, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)

for i in range(epochs):
    train_loss = train_loop(train_dl, model, loss_fn, optimizer, device)
    print(f'{i+1}번 epoch_loss: {train_loss}')

    model.save_pretrained(f'{DATA_PATH}models/skt/para_over_y15886->10000e30_ver.1_{i}')

    if train_loss < 0.16:
        break
````

````python
@torch.no_grad()
def chatbot(model, tokenizer, max_len, device):
    model.eval()
    while True:
        text = input('user >> ').strip()
        if text == 'qwer':
            break

        text = '<q>' + text + '</s><a>'
        x = tokenizer.encode(text, return_tensors='pt').to(device)
        result_ids = model.generate(x,
                            max_length=512, # 생성 최대 길이
                            repetition_penalty=2.0, # 반복 토큰 생성에 대한 패널티, 1보다 큰 수를 줘야함
                            use_cache=True, # 빠른 추론을 위한 캐시 여부
                            do_sample=True, # 확률적 샘플링
                            temperature=0.5, # 소프트맥스 온도 (1미만: 확률들의 편차가 높아짐 / 1초과: 확률들 편차 줄어듬)
                            top_k=20
                            )
        q_len = len(text) +1
        text = tokenizer.decode(result_ids[0])

        print('bot >> ', text[q_len:-4])
````

````python
chatbot(model, tokenizer, 512, device)
````

## Paraphrase 버전

````python
@torch.no_grad()
def chatbot_p(model, tokenizer, max_len, device):
    model.eval()
    while True:
        text = input('user >> ').strip()
        if text == 'qwer':
            break

        text = '<q>' + text + '</s><a>'
        x = tokenizer.encode(text, return_tensors='pt').to(device)
        result_ids = model.generate(x,
                            max_length=512, # 생성 최대 길이
                            repetition_penalty=2.0, # 반복 토큰 생성에 대한 패널티, 1보다 큰 수를 줘야함
                            use_cache=True, # 빠른 추론을 위한 캐시 여부
                            do_sample=True, # 확률적 샘플링
                            temperature=0.5, # 소프트맥스 온도 (1미만: 확률들의 편차가 높아짐 / 1초과: 확률들 편차 줄어듬)
                            top_k=20
                            )
        q_len = len(text) +1
        text = tokenizer.decode(result_ids[0])

        generated_text = pipe(text,
                             max_length = 128,
                             num_beams = 5,
                             num_return_sequences = 4,
                             no_repeat_ngram_size = 4,
                             top_k = 81,
                             top_p = 0.8288,
                             early_stopping=True)
        text = generated_text[0]['generated_text']

        print('bot >> ', text[q_len:-4])
````
